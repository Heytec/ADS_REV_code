```python
# import lib
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
%%time
# get the data 
df=pd.read_csv('https://raw.githubusercontent.com/anilak1978/ecommerce/master/OnlineRetail.csv',encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#check NAN
df.isnull().sum()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
df.head()
```

```python
df["Revenue"]=df["Quantity"]*df["UnitPrice"]
```

```python
df.head()
```

```python
df['InvoiceMonth']=pd.DatetimeIndex(df['InvoiceDate']).month
```

```python
df.head()
```

```python
df["InvoiceYear"]=pd.DatetimeIndex(df['InvoiceDate']).year
```

```python
df.head()
```

```python
df_revenue=df.groupby(["InvoiceMonth",'InvoiceYear'])['Revenue'].mean().reset_index()
df_revenue
```

```python
plt.figure(figsize=(15,10))

sns.barplot(x='InvoiceMonth',y='Revenue',data=df_revenue,hue='InvoiceYear')
plt.title("Monthly Revenue Overview")
plt.xlabel("Month")
plt.ylabel("Revenue")
```

```python
df_first_purchase=df.groupby(['CustomerID'])['InvoiceDate'].min().reset_index()
df_first_purchase.columns=["CustomerID", "FirstPurchaseDate"]

df_first_purchase
```

```python
df=pd.merge(df,df_first_purchase,on='CustomerID')
```

```python
df.head()
```

```python
df['UserType']='New'
```

```python
df.head()
```

```python
df.loc[df['InvoiceDate']>df['FirstPurchaseDate'],'UserType']="Existing"
```

```python
df.UserType.value_counts()
```

```python
df_new_revenue=df.groupby(["InvoiceMonth", "InvoiceYear", "UserType"])["Revenue"].sum().reset_index()
plt.figure()
sns.relplot(x="InvoiceMonth", y="Revenue", hue="UserType", data=df_new_revenue, kind="line", height=12, aspect=18/10)
plt.title("New vs Existing Customer Revenue Overview")
plt.xlabel("Month")
plt.ylabel("Revenue")
```
