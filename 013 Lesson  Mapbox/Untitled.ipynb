# Example 4 Mapbox using GPS data

```python
gps_data = "https://gist.githubusercontent.com/rahulbhadani/5f00dbc4a7db7cb4624b87ca0aad8a3e/raw/8faffb92790fb633e86f90e875ef0abf7c99a61c/gps.csv"
```

```python
gps_data=pd.read_csv(gps_data)
```

```python
gps_data.head()
```

```python
fig=px.scatter_mapbox(gps_data,
                      lat='Lat',
                      lon='Long',
                      color='Time',
                      color_continuous_scale=["black", "purple", "red" ], size_max=70, zoom=20,
                      height = 600, 
                      width = 1000, #center = dict(lat = g.center)
                      title='Drive Route with Mapbox',)


fig.update_layout(font_size=16, 
                  title={'xanchor': 'center','yanchor': 'top', 'y':0.9, 'x':0.5,}, 
                  title_font_size = 24, 
                  mapbox_accesstoken=mapbox_access_token, 
                  mapbox_style = "mapbox://styles/012345678909876543210/cjvgs71ls0ega1fqljye2mxdg")
```

```python
fig.show()
```

# Example 5 Counties of Kenya

```python
df = pd.read_excel('data/ke.xlsx')
df.head()
```

```python
fig=px.scatter_mapbox(df,lat='lat',lon='lng', hover_name="admin", hover_data=["city", "population"],zoom=10,width=900*1.5, height=500*1.5,color='city',
                            
                              size="population_proper",
                     size_max=70)

fig.update_layout(mapbox_style="mapbox://styles/012345678909876543210/ckkplul5j0kzj17lbkcilctz7")

fig.show()
```
